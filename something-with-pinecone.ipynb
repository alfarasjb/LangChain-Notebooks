```python
import os 
import warnings 
from dotenv import load_dotenv 

load_dotenv() 

PINECONE_API_KEY = os.getenv("PINECONE_API_KEY") 
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

warnings.filterwarnings('ignore')
```

```python
import pymupdf # imports the pymupdf library
def load_pdf_file(filename: str): 
    doc = pymupdf.open(filename)  # Open a document 
    doc_contents = []
    for page in doc:  # Iterate the document pages 
        text = page.get_text()  
        doc_contents.append(text) 
    doc_contents = "\n".join(doc_contents) 
    return doc_contents
doc_contents = load_pdf_file("data/AI integration.pdf")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=100, 
    chunk_overlap=0
)

texts = text_splitter.create_documents([doc_contents])
texts
```

```python
from langchain.embeddings import OpenAIEmbeddings 

embeddings = OpenAIEmbeddings() 
```

```python
import pinecone 
from langchain.vectorstores import Pinecone 

pinecone_client = pinecone.Pinecone(api_key=PINECONE_API_KEY) 
```

```python
index_name = "langchain-fed" 
search = Pinecone.from_documents(texts, embeddings, index_name=index_name)
```

```python
query = "What are the services offered?"
result = search.similarity_search(query) 
result
```
